```python
import openpyxl
workbook = openpyxl.Workbook()
worksheet = workbook.active

worksheet["B2"] = "請求書"
worksheet["B4"] = "株式会社ABC"
worksheet["B5"] = "〒101-0022 東京都千代田区神田練塀町300"
worksheet["B6"] = "TEL:03-1234-5678 FAX:03-1234-5678"
worksheet["B7"] = "担当者名:鈴木一郎 様"

worksheet["F4"] = "No."
worksheet["G4"] = "0001"
worksheet["F5"] = "日付"

import datetime
now = datetime.datetime.now()
date = datetime.datetime.strftime(now,"%Y/%m/%d")
worksheet["G5"] = date

start_row = 9
start_col = 2

header = ["商品名","数量","単価","金額"]

for col,title in enumerate(header, start=start_col):
  worksheet.cell(row=start_row, column=col, value = title)

data = [
    ["商品A",2,10000,20000],
    ["商品B",1,15000,15000]
]

for i,row_data in enumerate(data, start=start_row + 1):
  for j, value in enumerate(row_data, start=start_col):
    worksheet.cell(row=i, column=j, value=value)

total = sum(item[3] for item in data)
worksheet.cell(row=i+1, column=start_col+3, value=total)

worksheet.cell(row=i+3, column=start_col, value="合計")
worksheet.cell(row=i+3, column=start_col+3, value=total)

worksheet.cell(row=i+4, column=start_col, value="消費税")
worksheet.cell(row=i+4, column=start_col+3, value=total*0.1)

worksheet.cell(row=i+5, column=start_col, value="税込合計")
worksheet.cell(row=i+5, column=start_col+3, value=total*1.1)

date_filename = f"請求書_{now.strftime('%Y%m%d')}.xlsx"
workbook.save(date_filename)
```
