```python
import pandas as pd

df = pd.DataFrame({
    '日付' : ["2023/5/17","2023/5/18","2023/5/19","2023/5/20","2023/5/21"],
    '社員名' : ["山田","佐藤","鈴木","田中","高橋"],
    '売上' : [100,200,150,300,250],
    '部門' : ["メーカー","代理店","メーカー","商社","代理店"]
})

avg_sales = df["売上"].mean()
df["平均売上"]=avg_sales

def performance(rank):
  result = "";
  if rank > avg_sales:
    result = 'A';
  elif rank == avg_sales:
    result = 'B';
  else:
    result = 'C';
  return result

df["業績ランク"]=df["売上"].apply(performance)
writer = pd.ExcelWriter("業績.xlsx")

df.to_excel(writer,sheet_name="sheet1",index=False)
writer.close()
```
